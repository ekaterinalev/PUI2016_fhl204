## NULL HYPOTHESIS: 
The cumulative distribution of CitiBike riders' ages for pm riders is statistically similar to that of am riders, significance level p=0.05.
H0: R1 = R2

## ALTERNATE HYPOTHESIS: 
The cumulative distribution of CitiBike riders' ages for pm riders is statistically dissimilar than that of am riders, significance level p=0.05.

HA: R1 <> R2

## Results:  
statistic=0.030685849063871985, pvalue=5.9896064278657741e-53

If the K-S statistic is small or the p-value is high, then we cannot reject the hypothesis that the distributions of the two samples are the same.  In this case, we can reject the null hypothesis since the pvalue is below 1%.  Riders in the am and pm groups exhibit dissimilar age distributions.

```python
# my usual imports and setups
import pylab as pl
import pandas as pd
import numpy as np
#imports downloader
from getCitiBikeCSV import getCitiBikeCSV

%pylab inline

import os
import sys
import json

import datetime
import scipy.stats
import time
```

```python
if os.getenv("PUIDATA") is None:
    print ("must set PUIDATA env variable")
    sys.exit()

def getCitiBikeCSV(datestring):
    '''Downloads citibike data and unzips it. If the data is downloaded by not unzippeds it zips it. Moves the data to $PUIDATA
    Arguments:
        date string as yyyymm
    '''
    print ("Downloading", datestring)
    ### First I will heck that it is not already there
    if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.csv"):
        if os.path.isfile(datestring + "-citibike-tripdata.csv"):
            # if in the current dir just move it
            if os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA")):
                print ("Error moving file!, Please check!")
        #otherwise start looking for the zip file
        else:
            if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.zip"):
                if not os.path.isfile(datestring + "-citibike-tripdata.zip"):
                    os.system("curl -O https://s3.amazonaws.com/tripdata/" + datestring + "-citibike-tripdata.zip")
                ###  To move it I use the os.system() functions to run bash commands with arguments
                os.system("mv " + datestring + "-citibike-tripdata.zip " + os.getenv("PUIDATA"))
            ### unzip the csv 
            os.system("unzip " + os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.zip")
            ## NOTE: old csv citibike data had a different name structure. 
            if '2014' in datestring:
                os.system("mv " + datestring[:4] + '-' +  datestring[4:] + 
                          "\ -\ Citi\ Bike\ trip\ data.csv " + datestring + "-citibike-tripdata.csv")
            os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA"))
    ### One final check:
    if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.csv"):
        print ("WARNING!!! something is wrong: the file is not there!")

    else:
        print ("file in place, you can continue")
```

```python
datestring = '201501'
getCitiBikeCSV(datestring)
```

```python
df = pd.read_csv(os.getenv("PUIDATA") + "/" + datestring + '-citibike-tripdata.csv')
```

```python
df['date'] = pd.to_datetime(df['starttime'])
#df.head(n=5)
```

```python
df['TimeStr'] = df['date'].dt.strftime('%p')
#df.head(n=5)
```

```python
#By Time
df['ageam'] = 2016 - df['birth year'][(df['usertype'] == 'Subscriber') & (df['TimeStr'] == 'AM')]
df['agepm'] = 2016 - df['birth year'][(df['usertype'] == 'Subscriber') & (df['TimeStr'] == 'PM')]
```

```python
df['ageam'].dropna(inplace= True)
df['agepm'].dropna(inplace= True)
```

```python
bins = np.arange(10, 99, 5)
axam = df.ageam.groupby(pd.cut(df.ageam, bins)).agg([count_nonzero]).plot(kind='bar', 
                                                                legend=False)
axam.set_title("am riders")

axpm = df.agepm.groupby(pd.cut(df.agepm, bins)).agg([count_nonzero]).plot(kind='bar', 
                                                                legend=False)
axpm.set_title("pm riders")
```

```python
csam=df.ageam.groupby(pd.cut(df.ageam, bins)).agg([count_nonzero]).cumsum()

cspm=df.agepm.groupby(pd.cut(df.agepm, bins)).agg([count_nonzero]).cumsum()


print (np.abs(csam / csam.max() - cspm / cspm.max()))

pl.plot(bins[:-1] + 5, csam / csam.max(), label = "M")
pl.plot(bins[:-1] + 5, cspm / cspm.max(), label = "F")

pl.plot(bins[:-1] + 5, np.sqrt(cspm / cspm.max() - csam / csam.max())**2, 'k-',
        label = "difference")
pl.xlabel("Age")
pl.ylabel("Normalized Cumulative Number")
pl.legend()
```

```python
ks = scipy.stats.ks_2samp(df.ageam, df.agepm)
print (ks)
```
